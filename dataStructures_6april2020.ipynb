# Data structures

## Melissa

$$\textrm{Richness} \sim \textrm{species} + \textrm{time} + \textrm{treatment:prepost} + (1| \textrm{individual} ) + (1 | \textrm{time})$$

$$R_i \sim N(X_i \beta , \sigma_i)$$

Individual-level predictors:

$$X = \left[\begin{array} 
{rrrrrrrrrr}
&\textrm{intercept (species1, con:pre)} & \textrm{species2} & \textrm{species3} & \textrm{species4} & \textrm{species5} & \textrm{time} & \textrm{con:post} & \textrm{treat:pre} & \textrm{treat:post}  \\
\textrm{Sample_1}& 1& X_{[1,sp2]} & X_{[1,sp3]} & X_{[1,sp4]} & X_{[1,sp5]} & X_{[1,"time"]} & X_{[1,conpost]} & X_{[1,treatpre]} &  X_{[1,treatpost]}\\
\textrm{Sample_2}& 1& X_{[2,sp2]} & X_{[2,sp3]} & X_{[2,sp4]} & X_{[2,sp5]} & X_{[2,"time"]} & X_{[2,conpost]} & X_{[2,treatpre]} &  X_{[2,treatpost]}\\
\vdots & & &  &  \ddots &  &  &  & &  \vdots\\
\textrm{Sample_n}& 1& X_{[n,sp2]} & X_{[n,sp3]} & X_{[n,sp4]} & X_{[n,sp5]} & X_{[n,"time"]} & X_{[n,conpost]} & X_{[n,treatpre]} &  X_{[n,treatpost]}
\end{array}\right]$$

Individual-level coefficients
$$\beta = \left[ \begin{array}{rr}
\textrm{intercept}& \alpha + \alpha_{I[i]} + \alpha_{T[i]} \\
\textrm{species2}& \beta_1\\
\textrm{species3}& \beta_2\\
\textrm{species4}& \beta_3\\
\textrm{species5}& \beta_4\\
\textrm{speciestime}& \beta_5\\
\textrm{con:post}& \beta_6\\
\textrm{treat:pre}& \beta_7\\
\textrm{treat:post}& \beta_8
\end{array}\right]$$

$$\alpha_I \sim N(0, \sigma_I)$$
$$\alpha_T \sim N(0, \sigma_T)$$





## Courtney

$$\textrm{phenology} \sim \textrm{treatment} + \textrm{siteT}  + \textrm{yearT} + \textrm{treatment:siteT} + \textrm{treatment:yearT} + (\textrm{treatment}|\textrm{species}) + (\textrm{treatment}|\textrm{site:year}) + (\textrm{treatment}|\textrm{site:subsite}) $$


$$P_{sp[i],ss[i],y[i]} \sim N(X_i \beta , \sigma_i)$$

and there are $k$ species, $l$ years, and $m$ subsites.


Individual-level predictors:

$$X = \left[\begin{array} 
{rrrrrrr}
&\textrm{intercept (Con)} & \textrm{Treat} & \textrm{siteT} & \textrm{yearT} & \textrm{Treat:siteT} & \textrm{Treat:yearT}   \\
\textrm{PhenologyIntercept_1} & 1 & X_{[1,"Treat"]} & X_{[1,"siteT"]} & X_{[1,"yearT"]} & X_{[1,"siteT:Treat"]} & X_{[1,"yearT:Treat"]}   \\
\textrm{PhenologyIntercept_2} & 1 & X_{[2,"Treat"]}  & X_{[2,"siteT"]} & X_{[2,"yearT"]} & X_{[2,"siteT:Treat"]} & X_{[2,"yearT:Treat"]}   \\
\vdots &\vdots&\vdots&\vdots&\vdots&\vdots&\vdots&\vdots\\
\textrm{PhenologyIntercept_n} & 1 & X_{[n,"Treat"]}  & X_{[n,"siteT"]} & X_{[n,"yearT"]} & X_{[n,"siteT:Treat"]} & X_{[n,"yearT:Treat"]}   \\
\end{array}\right]$$

Individual-level coefficents:

Below, $S$ is species with $k$ different species, $Y$ is site:year with $l$ different site:year combos, and $B$ is site:subsite with $m$ different site:subsite combos. Therefore,

$$\beta_{S[i]} =  \left[\begin{array} 
{rrrrrrrrr}
&\textrm{intercept (Con)} & \textrm{Treat}  & \textrm{siteT} & \textrm{yearT} & \textrm{Treat:siteT} & \textrm{Treat:yearT} \\
S=1,Y=1,B=1 &\alpha +  \alpha_{S[1]} + \alpha_{Y[1]} + \alpha_{B[1]}  & \beta_{S[1]} + \beta_{Y[1]} + \beta_{B[1]} & \beta_2 & \beta_3 & \beta_4 & \beta_6 & \beta_7 \\
S=1,Y=1,B=2 & \alpha + \alpha_{S[2]} + \alpha_{Y[2]} + \alpha_{B[2]} & \beta_{S[2]} + \beta_{Y[2]} + \beta_{B[2]}& \beta_2 & \beta_3 & \beta_4 & \beta_6 & \beta_7  \\
\vdots &\vdots&\vdots&\vdots&\vdots&\vdots&\vdots&\vdots&\vdots\\
S=k, Y=l, B=m &  \alpha + \alpha_{S[k]} + \alpha_{Y[l]} + \alpha_{B[m]} & \beta_{S[k]} + \beta_{Y[l]} + \beta_{B[m]}  & \beta_2 & \beta_3 & \beta_4 & \beta_6 & \beta_7  \\
\end{array}\right]$$

$$\beta_S \sim N(0, \sigma^\beta_S)$$
$$\beta_Y \sim N(0, \sigma^\beta_Y)$$
$$\beta_B \sim N(0, \sigma^\beta_B)$$

$$\alpha_S \sim N(0, \sigma^\alpha_S)$$
$$\alpha_Y \sim N(0, \sigma^\alpha_Y)$$
$$\alpha_B \sim N(0, \sigma^\alpha_B)$$



## Anna

Trap-level variables: collection day of year, LAI

Plot-level: nlcd Class

Site-level: degree-days, recent temp

$$\textrm{abundance} \sim \textrm{collectionDate} + \textrm{LAI} + \textrm{degreeDay} + \textrm{recentTemp} + \textrm{nlcdClass} + (1|\textrm{plot}) + (1|\textrm{plot:trap})$$

$$A_i \sim N(X_i \beta_{T[i]} , \sigma_i)$$


Individual-level predictors:

$$X = \left[\begin{array} 
{rrrrr}
&\textrm{intercept} & \textrm{collectionDate} & \textrm{degreeDays} & \textrm{recentTemp}\\
\textrm{Sample_1}& 1 & X_{[1,"collectionDate"]} & X_{[1,"degreeDays"]} & X_{[1,"recentTemp"]}\\
\textrm{Sample_2}& 1& X_{[2,"collectionDate"]} & X_{[2,"degreeDays"]} & X_{[2,"recentTemp"]}\\
\vdots &\vdots &  \vdots &  \vdots&\vdots \\
\textrm{Sample_n}& 1& X_{[n,"collectionDate"]} & X_{[n,"degreeDays"]} & X_{[n,"recentTemp"]}
\end{array}\right]$$

Individual-level coefficents:
$$\beta_{T[i]} = \left[ \begin{array}{rrrrr} 
& \textrm{intercept} &\textrm{collectionDate} & \textrm{degreeDays} & \textrm{recentTemp} \\
\textrm{Trap_1} & \alpha_{T[1]}& \beta_{DOY} & \beta_{degreeDays}  & \beta_{recentTemp} \\
\textrm{Trap_2} & \alpha_{T[2]}& \beta_{DOY} & \beta_{degreeDays}  & \beta_{recentTemp}\\
\vdots & & & \vdots \\
\textrm{Trap_k} & \alpha_{T[k]}& \beta_{DOY} & \beta_{degreeDays}  & \beta_{recentTemp}\\
\end{array}\right]$$

$$\alpha_T =\sim N(U_T G^T, \sigma_T)$$

#### Trap-level

Group-level predictors  [ Trap ]
$$U_T = \left[ \begin{array}{rrr} 
& \textrm{Intercept} & \textrm{LAI}\\
\textrm{Trap_1} & 1 & u^T_{[1,\textrm{LAI}]}\\
\textrm{Trap_2} & 1 &  u^T_{[2,\textrm{LAI}]}\\
\vdots & \\
\textrm{Trap_k} & 1 & u^T_{[k,\textrm{LAI}]}\\
\end{array}\right]$$

Group-level coefficients  [ Trap ]
$$ G^T = \left[ \begin{array}{rrr} 
 & \textrm{intercept} \\
\textrm{Intercept} & \alpha_{P[k]}^0 \\
\textrm{Effect of LAI} & \gamma_{LAI}^0 \\
\end{array}\right]$$

#### Plot-level

$$ \alpha_P \sim N(U_P G^P, \sigma_P)$$

Second-Group-level predictors  [ Plot ]
$$U_P = \left[ \begin{array}{rrr} 
& \textrm{Intercept} & \textrm{nlcdClass}\\
\textrm{Plot_1} & 1 & u^P_{[1,\textrm{nlcdClass}]}\\
\textrm{Plot_2} & 1 &  u^P_{[2,\textrm{nlcdClass}]}\\
\vdots & \\
\textrm{Plot_L} & 1 & u^P_{[L,\textrm{nlcdClass}]}\\
\end{array}\right]$$

Second-group-level coefficients  [ Plot ]
$$ G^P = \left[ \begin{array}{rrr} 
 & \textrm{intercept} \\
\textrm{Intercept} & \gamma^0 \\
\textrm{Effect of nlcdClass} & \gamma_{LAI}^0 \\
\end{array}\right]$$
